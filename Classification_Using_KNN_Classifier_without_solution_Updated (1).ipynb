## üìÑ Hypothyroid Dataset

This dataset is used to help diagnose thyroid-related conditions based on lab tests, medical history, and patient information.

---
- **age**: Age of the patient  
- **sex**: Gender of the patient (`0.0` , `1.0`)
- **on thyroxine**: Whether the patient is on thyroxine medication (`0` or `1`)  
- **query on thyroxine**: Whether the patient's thyroxine status was queried  (`0` or `1`)
- **on antithyroid medication**: Whether the patient is taking antithyroid medication  (`0` or `1`)
- **sick**: Whether the patient is currently sick  (`0` or `1`)
- **pregnant**: Whether the patient is pregnant  (`0` or `1`)
- **thyroid surgery**: Whether the patient has had thyroid surgery  (`0` or `1`)
- **I131 treatment**: Whether the patient has undergone I131 radioactive treatment  (`0` or `1`)
- **query hypothyroid**: Whether hypothyroidism was queried  (`0` or `1`)
- **query hyperthyroid**: Whether hyperthyroidism was queried  (`0` or `1`)
- **lithium**: Whether the patient is taking lithium  (`0` or `1`)
- **goitre**: Presence of a goitre  (`0` or `1`)
- **tumor**: Presence of a tumor  (`0` or `1`)
- **hypopituitary**: Presence of hypopituitarism  (`0` or `1`)
- **psych**: Presence of a psychological disorder(`0` or `1`)
- **TSH measured**: Indicates if the TSH test was performed  (`0` or `1`)
- **TSH**: Thyroid Stimulating Hormone test result (numeric)
- **T3 measured**: Indicates if the T3 test was performed  (`0` or `1`)
- **T3**: Triiodothyronine test result (numeric)
- **TT4 measured**: Indicates if the TT4 test was performed  (`0` or `1`)
- **TT4**: Total Thyroxine test result (numeric)  
- **T4U measured**: Indicates if the T4U test was performed  (`0` or `1`)
- **T4U**: Thyroxine Uptake test result (numeric)  
- **FTI measured**: Indicates if the Free Thyroxine Index was measured  (`0` or `1`)
- **FTI**: Free Thyroxine Index result (numeric)  
- **TBG measured**: Indicates if the TBG test was performed  (`0` or `1`)
- **TBG**: Thyroxine Binding Globulin test result (numeric) - `NaN values
- **referral source**: The source from which the patient was referred (e.g., SVHC, other, SVI, etc.) but encoded with frequency encoder.

---

### üè∑Ô∏è Target Label
- **binaryClass**: Final diagnosis/classification ( `N` or `P`) encoded as (`0` or `1`) respectively.

### 8- Split the data to 70% training data and 30% testing data and display their sizes: You can refer to <a href = "https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html"><u>sklearn documntation</u></a> that contains an example.

```python
## Import the train_test_split from sklearn library:
from sklearn.model_selection import train_test_split

## Split into 70% for training and %30 testing:
X = df.drop(columns=["binaryClass"])
y = df["binaryClass"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

## Display the shapes for X_train:
print("Shape of X_train: ", X_train.shape)
## Display the shapes for X_test:
print("Shape of X_test: ", X_test.shape)
## Display the shapes for y_train:
print("Shape of y_train: ", y_train.shape)
## Display the shapes for y_test:
print("Shape of y_test: ", y_test.shape)
```

### 10- Apply Standardization:

We can implement it using <a href= "https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html"><u>sklearn</u></a>.

```python
## Import StandardScaler:
from sklearn.preprocessing import StandardScaler

## Initialize the StandardScaler
scaler = StandardScaler()

## Fit the scaler to X_train , X_test and transform it:
scaler.fit(X_train)
X_train = scaler.transform(X_train)
scaler.fit(X_test)
X_test = scaler.transform(X_test)



## Display the shapes for X_train:
print("Shape of X_train after scaling: ", X_train.shape)
## Display the shapes for X_test:
print("Shape of X_test after scaling: ", X_test.shape)
## Display the shapes for y_train:
print("Shape of y_train after scaling: ", y_train.shape)

## Display the shapes for y_test:
print("Shape of y_test after scaling: ", y_test.shape)
```

### 2- Determine the optimal value of k for the K-Nearest Neighbors (KNN) classifier by evaluating its performance across a range of k values.  Plot the classification accuracy for each k value to visualize how model performance changes, and identify the k value that yields the highest accuracy.:
<hr>

```python
## Import the libraries you need (Accuracy and KNN):
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

```python
# Store accuracy for each k
accuracy_scores = []

# Range of k to try
k = range(1,29,2)
for i in k:
    ## Initialize KNN:
    knn = KNeighborsClassifier(n_neighbors=i)

    ## Train with scaled training features and training target
    knn.fit(X_train, y_train)

    ## Get the predicted values after training:
    y_pred = knn.predict(X_test)

    ## Find the accuracy score:
    accuracy = accuracy_score(y_test, y_pred)

    ## Add the accuracy score to the list
    accuracy_scores.append(accuracy)

    ## Display the obtained accuracy:
    print(f"Accuracy for k={i}: {accuracy:.2f}")

# Find the best k and display it
print(f"Best k: {k[accuracy_scores.index(max(accuracy_scores))]}")


#  Plot the results
plt.plot(k, accuracy_scores, marker='o')
plt.title('KNN Classifier Accuracy vs. K Value')
plt.xlabel('K Value')
plt.ylabel('Accuracy')
plt.xticks(k)
plt.grid()
plt.show()
```

### 3- Evaluate KNN using different evaluation metrics (Precision, Recall, F1 score and Accuracy) :
<hr>

```python
## Import all the needed evaluation metrics:
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

## Initialize knn
knn = KNeighborsClassifier(n_neighbors=9)


## Fit knn with the selected features:
knn.fit(X_train, y_train)

## Predict the values from knn:
y_pred = knn.predict(X_test)

# Calculate Accuracy:
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.5f}")


# Calculate Precision
precision = precision_score(y_test, y_pred)
print(f"Precision: {precision:.5f}")


# Calculate Recall
recall = recall_score(y_test, y_pred)
print(f"Recall: {recall:.5f}")


# Calculate F1-Score
f1 = f1_score(y_test, y_pred)
print(f"F1-Score: {f1:.5f}")
```
